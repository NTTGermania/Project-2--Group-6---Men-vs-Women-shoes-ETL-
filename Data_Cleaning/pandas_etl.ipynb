```python
import pandas as pd
from sqlalchemy import create_engine, inspect
```

### Extract CSVs into DataFrames

```python
women_shoe_file = "../Resources/7003_1.csv"
women_shoe_df = pd.read_csv(women_shoe_file)
women_shoe_df.head()
```

```python
men_shoe_file = "../Resources/7004_1.csv"
men_shoe_df = pd.read_csv(men_shoe_file)
men_shoe_df.head()
```

```python
women_shoe_df["categories"]
```

```python
women_shoe_df["category"] = "Women"
women_shoe_df.head()
```

```python
men_shoe_df["categories"]
```

```python
men_shoe_df["category"] = "Men"
men_shoe_df.head()
```

### Transform women's shoes DataFrame

```python
# Create a filtered dataframe from specific columns
women_shoe_cols = ["id", "brand", "category", "prices_amountmin", "prices_amountmax"]
women_shoe_transformed = women_shoe_df[women_shoe_cols].copy()

# Rename the column headers
women_shoe_transformed = women_shoe_transformed.rename(columns={"prices_amountmin": "($)prices_min",
                                                          "prices_amountmax": "($)prices_max"})

# Clean the data by dropping duplicates and setting the index
women_shoe_transformed.drop_duplicates("id", inplace=True)
women_shoe_transformed.set_index("id", inplace=True)

women_shoe_transformed.head()
```

# Transform men's shoes DataFrame

```python
# Create a filtered dataframe from specific columns
men_shoe_cols = ["id", "brand", "category", "prices_amountmin", "prices_amountmax"]
men_shoe_transformed = men_shoe_df[men_shoe_cols].copy()

# Rename the column headers
men_shoe_transformed = men_shoe_transformed.rename(columns={"prices_amountmin": "($)prices_min",
                                                          "prices_amountmax": "($)prices_max"})

# Clean the data by dropping duplicates and setting the index
men_shoe_transformed.drop_duplicates("id", inplace=True)
men_shoe_transformed.set_index("id", inplace=True)

men_shoe_transformed.head()
```

### Create database connection

```python
protocol = 'postgresql'
username = 'postgres'
password = 'password'
host = 'localhost'
port = 5432
database_name = 'women_shoe_transformed'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
# Confirm tables
inspector = inspect(engine)
inspector.get_table_names()
```

### Load DataFrames into database

```python
women_shoe_transformed.to_sql(name='women_shoe_transformed', con=engine, if_exists='append', index=True)
```

```python
men_shoe_transformed.to_sql(name='men_shoe_transformed', con=engine, if_exists='append', index=True)
```
